```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from hmmlearn import hmm

states = ["dificil", "mediu", "usor"]
observations = ["FB", "B", "S", "NS"]
n_states = len(states)
n_observations = len(observations)

obs_sequence = np.array([observations.index(obs) for obs in ["FB", "FB", "S", "B", "B", "S", "B", "B", "NS", "B", "B", "S"]]).reshape(-1, 1)

#print(obs_sequence)
transition_matrix = np.array([
    [0, 0.5, 0.5],
    [0.5, 0.25, 0.25],
    [0.5, 0.25, 0.25]
])

start_probabilities = np.array([1/3, 1/3, 1/3])

emission_matrix = np.array([
    [0.1, 0.2, 0.4, 0.3],
    [0.15, 0.25, 0.5, 0.1],
    [0.2, 0.3, 0.4, 0.1]
])


model = hmm.CategoricalHMM(n_components=n_states)
model.startprob_ = start_probabilities
model.transmat_ = transition_matrix
model.emissionprob_ = emission_matrix

hidden_states = model.predict(obs_sequence)
print("secventa de stari ascunse:", hidden_states)



sns.set_style("darkgrid")
plt.plot(hidden_states, '-o', label="Hidden State")
plt.xlabel("Time Step")
plt.ylabel("Hidden State (Word or Silence)")
plt.yticks(ticks=range(n_states), labels=states)
plt.legend()
plt.title("Predicted Hidden States Over Time")
plt.show()
```
