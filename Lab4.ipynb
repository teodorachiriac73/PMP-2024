```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (3, 4), (4, 5)]
G.add_edges_from(edges)

plt.figure()
nx.draw(G, with_labels=True, node_color='lightblue', node_size=1000, font_size=10, font_weight='bold', edge_color='gray')
plt.title("Graful reelei Markov")
plt.show()
```

```python
cliques = list(nx.find_cliques(G))

print("Clicile:")
for clique in cliques:
    print(clique)
```

```python
import itertools
import numpy as np

def energie(clica):
    sum = 0
    for i in range(len(clica)):
      sum += (i + 1) * clica[i]
    return sum

clici=[[1, 0],
[1, 3, 4],
[2, 0],
[2, 3]] #am scazut cu 1 din noduri ca sa se potriveasca cu indicii vectorului
toateCombinatiile = list(itertools.product([0,1],repeat=5))
print(toateCombinatiile) #toate combinatiile posibile A0,A1,A2,A3,A4

Z = sum(energie(combinatie) for combinatie in toateCombinatiile)
print('Z=' + str(Z))


probabilitati=[]
for combinatie in toateCombinatiile:

  #print('combinatie '+ str(combinatie))
  A = []
  for index, value in enumerate(combinatie):
      A.append(value)
  for clica in clici:
    valoare=energie(clica)/Z

    probabilitati.append(valoare)

for clica_prob in probabilitati:
    print(clica_prob)

stari_prob_maxima = max(probabilitati)
print('probabilitatea maxima')
print(stari_prob_maxima)
```
