Urna in care avem 3 bile rosii, 4 albastre si 2 negre

 Aruncam un zar; daca obtinem un numar prim, adaugam o bila neagra in urna, daca obtinem 6, adaugam o bila rosie, iar in celelalte cazuri adaugam o bila albastra. Apoi extragem o bila din urna. 

 #### Rezolvare
a) Simulam evenimentul prin a genera o valoare random a zarului.

```python
import numpy as np 

urna=[3,4,2]
culoriBile=['rosu','albastru','negru']


zar= np.random.randint(1,7)
print(zar)
if(zar==2 or zar==3 or zar==5): 
    print('adaug bila neagra')
    urna[2]=urna[2]+1
elif(zar==6): 
    print('adaug bila rosie')
    urna[0]=urna[0]+1
else: 
    print('adaug bila albastra')
    urna[1]=urna[1]+1
#am adaugat o bila in functie de zar 
print(" urna dupa adaugare bila "+ str(urna))
#apoi extrag o bila
culoare= np.random.choice(culoriBile)
if(culoare=='rosu'): urna[0]-=1
elif(culoare=='albastru'): urna[1]-=1
elif(culoare=='negru'): urna[2]-=1
print(" urna dupa extragere bila "+ culoare +" "+ str(urna))
```

b) Estimati probabilitatea de a obtine o bila rosie.

```python
import numpy as np

def functie(nr_simulari):
    nr_rosii=0
    for _ in range (nr_simulari):
        urna=[3,4,2]
        
        culoriBile=['rosu','albastru','negru']
        zar= np.random.randint(1,7)
        print(zar)
        if(zar==2 or zar==3 or zar==5): 
            print('adaug bila neagra')
            urna[2]=urna[2]+1
        elif(zar==6): 
            print('adaug bila rosie')
            urna[0]=urna[0]+1
        else: 
            print('adaug bila albastra')
            urna[1]=urna[1]+1
        #am adaugat o bila in functie de zar 
        print(" urna dupa adaugare bila "+ str(urna))
        #apoi extrag o bila
        culoare= np.random.choice(culoriBile)
        if(culoare=='rosu'): 
            urna[0]-=1
            nr_rosii+=1
        elif(culoare=='albastru'): urna[1]-=1
        elif(culoare=='negru'): urna[2]-=1

    return nr_rosii


nr_incercari=10000
rezultat=functie(nr_incercari) 
print("probabilitate de a obtine rosie ")
print(rezultat/nr_incercari)
```
